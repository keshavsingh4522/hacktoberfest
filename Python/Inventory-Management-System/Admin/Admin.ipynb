#INVENTORY Management System

## Admin part 
  Admin works with main repository where he can add items , delete items and update items. finally see the sales status.


### Admin Options
- Add items
- Delete items 
- Update items count
- View Inventory
- View Sales
- View Pre bookings
- exit

```python
import json
  
```

```python
def add(obj):
  ID = obj.keys()
  #print(int(max(ID))+1)
  #{"1000":{"name":"Note Book","Cost": 100 ,"qt":20,"seller":"classmate","availability":"YES","PreBooking":0 }}
  name = input("Enter the product Name: ")
  Cost = int(input("Enter the product Cost: "))
  qt = int(input("Enter the product Quantity: "))
  seller = input("Enter the product Seller Name: ")
  availability = "YES"
  PreBooking = 0
  obj[str(int(max(ID))+1)] = {"name": name,"Cost": Cost ,"qt": qt,"seller":seller,"availability" :availability ,"PreBooking":0 }
  json_object1 = json.dumps(obj)
# Writing to sample.json
  with open("repository.json", "w") as outfile:
      outfile.write(json_object1)
  return
```

```python
def det(obj):
  obj.pop(input("Enter Item ID:"))
  json_object1 = json.dumps(obj)
  with open("repository.json", "w") as outfile3:
    outfile3.write(json_object1)
  return
```

```python
def upd(obj):
  ID = input("Enter Item ID:")
  gt = int(input("Enter Quantity to be added: "))
  avail = 'YES'
  Pre = 0
  if gt >= obj[ID]["PreBooking"]:
    obj[ID]["qt"] = obj[ID]["qt"] + gt
    obj[ID]["availability"] = avail
    obj[ID]["PreBooking"] = Pre
  else:
    print("Quantity added is less than pre order cross validate:")
  json_object1 = json.dumps(obj)
  with open("repository.json", "w") as outfile:
    outfile.write(json_object1)
  return
```

```python
def show(obj):
  print(json.dumps(obj, indent = 1))
  return
```

```python
def sales():
  with open('sales.json', 'r') as openfile1:  
    json_object2 = json.load(openfile1)
  print(json.dumps(json_object2, indent = 1))
```

```python
def Dispre(obj):
  for i in obj.keys():
    if obj[i]["PreBooking"] > 0:
      print(i + " : " , end = " " )
      print(obj[i])
      print()
  return
```

```python
#driver code
print("Welcome Admin")
with open('repository.json', 'r') as openfile:
  json_object = json.load(openfile)
flag = 1

while(flag):
  print("Enter the option to preform operation\n1: Add Item \n2:Delete Item  \n3:Update Item \n4:View Inventory \n5:View Sales \n6:Check Pre Orders\n7:Exit")
  print()
  opt = int(input("Enter option: "))
  if opt ==1:
    add(json_object)
  elif opt == 2:
    det(json_object)
  elif opt == 3:
    upd(json_object)
  elif opt == 4:
    show(json_object)
  elif opt == 5:
    sales()
  elif opt == 6:
    Dispre(json_object)
  else:
    print("Thank You")

    flag = 0
  print("____________________________________________________________________________________________________")
  print()
```
