```python
from bs4 import BeautifulSoup
import requests
```

```python
page = 1

title, price, review = [],[],[]

headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"}

main_url = "https://www.nykaa.com"

while page<3:
    
    print(f"Page : {page}".center(50,'-'))
    print("\n\n")
    
    response = requests.get(f"https://www.nykaa.com/skin/cleansers/c/8378?page_no={page}&sort=popularity&root=nav_2&eq=desktop",headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    product_list = soup.find('div',attrs={'id':'product-list-wrap'}).find_all('div',attrs={'class':'productWrapper'})
    
    for i in range(len(product_list)):
        
        print(f"{i}".center(50,"*"))
        print(main_url+product_list[i].find('div',class_="css-d5z3ro").find('a')['href'])
        
        soup1 = BeautifulSoup(requests.get(main_url+product_list[i].find('div',class_="css-d5z3ro").find('a')['href'],headers=headers).content, 'html.parser')
        
        title.append(soup1.find('h1',class_="css-1gc4x7i").text)  #Name of product
        price.append("".join(list(soup1.find('span',class_="css-1jczs19").text)[1:])) #Price of product
        review.append(soup1.find_all('div',class_="css-1hvvm95")[-1].text.split()[0]) #Total reviews of product
        
        print(f"""
            Title : {title[-1]}
            Price : {price[-1]}
            Review: {review[-1]}
        """)
    page+=1
```

```python
data = {
    "Title" : title,
    "Price(INR)" : price,
    "Review" : review
       }
```

```python
from pandas import DataFrame
df = DataFrame(data)
```
