```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
X=np.arange(10)
Y=(X-5)**2
print(X)
print(Y)
```

```python
plt.style.use('seaborn')
plt.xlabel('X')
plt.ylabel('F(x)')
plt.plot(X,Y)
plt.show()
```

### Finding Value of X that minimises Y which is F(x)

```python
x=0
lr=0.1   #step size
error=[]
plt.plot(X,Y)
for i in range (50):
    gradient=2*(x-5)
    x=x-lr*gradient #downhill direction
    y=(x-5)**2
    error.append(y)
    plt.scatter(x,y)
    print(x)    
```
