```python
from sklearn.datasets import make_regression
import numpy as np
import matplotlib.pyplot as plt
```

```python
X,Y=make_regression(n_samples=400,n_informative=1,n_features=1,noise=1.8,random_state=11)
print(X.shape,Y.shape)
```

```python
Y=Y.reshape((-1,1))
print(Y)
```

```python
#Normalisation
X=(X-X.mean())/X.std()
print(X)
```

```python
plt.style.use('seaborn')
plt.figure()
plt.scatter(X,Y)
plt.show()
```

```python
#adding x0 as 1 in every coloumn

ones=np.ones((X.shape[0],1))
x_=np.hstack((X,ones))
print(x_.shape)
```

```python
def prediction(X,theta):
    return np.dot(X,theta)
def getTheta_ClosedFormSolution(X,Y):
    
    Y=np.mat(Y)
    first_part=np.dot(X.T,X)
    second_part=np.dot(X.T,Y)
    
    theta=np.linalg.pinv(first_part)*second_part
    return theta
```

```python
theta=getTheta_ClosedFormSolution(x_,Y)
print(theta)
```

```python
plt.style.use('seaborn')
plt.figure()
plt.scatter(X,Y)
plt.plot(X,prediction(x_,theta),color='red')
plt.show()
```
